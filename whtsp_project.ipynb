```python
import pandas as pd
import re
```

```python
f = open('chat.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = re.compile(r'(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[ap]m\s-\s)')
```

```python
messages = re.split(pattern, data)[1:]
len(messages)
```

```python
split_data = re.split(pattern, data)
# The split_data list will alternate between date strings and message strings.
# We need to pair the dates with the messages that follow them.
# The first element might be an empty string if the data starts with a date pattern.
messages = [split_data[i] for i in range(2, len(split_data), 2)]
len(messages)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%y, %I:%M %p - ')
```

```python
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```
